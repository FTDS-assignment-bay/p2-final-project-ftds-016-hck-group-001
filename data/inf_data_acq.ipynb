```python
import pandas as pd
```

```python
# cleaned
df_1 = pd.read_csv('consumer_complaints_cleaned.csv')

# labeled
df_2 = pd.read_csv('consumer_complaints_labeled.csv')

# combine 2 df
df_combined = pd.concat([df_1, df_2], axis=1)
df_combined

# change data type
df_combined['date_sent_to_company'] = pd.to_datetime(df_combined['date_sent_to_company'], format='%Y-%m-%d')
df_combined['date_received'] = pd.to_datetime(df_combined['date_received'], format='%Y-%m-%d')
```

```python
df_combined
```

```python
df_inf = df_combined[~df_combined[['product', 'sub_product', 'issue', 'sub_issue', 'consumer_complaint_narrative']].duplicated()].sample(5)
df_inf
```

```python
df_inf.to_csv('consumer_complaints_inf.csv', sep=',', index=False)
```
